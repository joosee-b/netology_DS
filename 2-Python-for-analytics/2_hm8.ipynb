#### Задание 1. Рассчитать и визуализировать корреляции для переменных данного датасета

```python
import random
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import kurtosis
from scipy.stats import skew

from pylab import rcParams

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline  


wine_data = pd.read_csv("wine_data.csv")


corr = wine_data.corr()

rcParams['figure.figsize'] = 15, 10
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

#### Задание 2. Визуализировать взаимосвязи между 3-4 переменными

```python
wine_data = pd.read_csv("wine_data.csv")
wine_data = wine_data.sort_values(by = 'Alcohol', ascending = False)

alc = [i for i in wine_data['Alcohol']]
malic = [i for i in wine_data['Malic acid']]

plt.plot(alc, malic)
plt.xlabel('Alcohol')
plt.ylabel('Malic acid')
```

```python
wine_data = wine_data.sort_values(by = 'Alcohol', ascending = False)

alc = [i for i in wine_data['Alcohol']]
mag = [i for i in wine_data['Magnesium']]

plt.plot(alc, mag)
plt.xlabel('Alcohol')
plt.ylabel('Magnesium')
```
